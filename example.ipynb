### Imports

```python
import numpy as np
import FillLevel
from matplotlib import style
style.use('bmh')
```

### Load the geometry, build functions, make plots

```python
fl = FillLevel.FillLevel('geometries/PurityMonitor.yaml')
fl.build_fill_funcs()

# mass filled in grams given a fill level in cm. Input can be a number or an array
mass = fl.mass_filled(16.7)
print('Mass filled: {:.1f} g'.format(mass))

# fill level in cm given a mass filled in grams. Input can be a number or an array
level = fl.fill_level(2900)
print('Fill level: {:.1f} cm'.format(level))

# make the plots
fl.plot_fill_level(np.linspace(0,4000,10000))
fl.plot_mass_filled(np.linspace(0,30,10000))
fl.plot_geometry()
```
